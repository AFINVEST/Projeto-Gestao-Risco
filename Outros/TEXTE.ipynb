```python
def calculate_portfolio_risk(df_returns, assets, weights):
    # Seleção dos retornos dos ativos
    df_portifolio = df_returns[assets].dropna().tail(252)  # Últimos 252 dias úteis
    df_portifolio

    # Passo 1: Calcular a volatilidade (desvio padrão anualizado) de cada ativo
    vol_anual = df_portifolio.std() * np.sqrt(252)
    print(vol_anual)

    # Passo 2: Calcular a matriz de covariância anualizada
    cov_matrix = df_portifolio.cov() * 252

    print(cov_matrix)

    # Passo 3: Calcular a volatilidade vezes o peso de cada ativo
    vol_ponderada = vol_anual * weights

    print(vol_ponderada)

    # Passo 4: Elevar a volatilidade ponderada ao quadrado e somar
    risco_individual = np.sum((vol_ponderada ** 2))

    print(risco_individual)
    
    # Adicionar o termo de covariância
    num_assets = len(assets)
    risco_covariancia = 0

    for i in range(num_assets):
        for j in range(i + 1, num_assets):
            risco_covariancia += 2 * cov_matrix.iloc[i, j] * weights[i] * weights[j]

    # Soma total do risco
    risco_total = risco_individual + risco_covariancia

    # Passo 5: Tirar a raiz quadrada para obter o risco do portfólio
    risco_portifolio = np.sqrt(risco_total)

    return risco_portifolio
```

```python
#Def analise do VAR do portifolio
def calculate_portfolioVaR(df_returns, assets, weights):
    # Seleção dos retornos dos ativos
    df_portifolio = df_returns[assets].dropna().tail(252)  # Últimos 252 dias úteis
    df_portifolio

    # Passo 1: Calcular a var (5% percentil) de cada ativo
    var = df_portifolio.quantile(0.05)
    print(var)


    # Passo 2: Calcular a matriz de correlação anualizada
    cor_matrix = df_portifolio.corr()
    

    print(cor_matrix)


    # Passo 3: Calcular o quadrado do Var ponderada e do Peso

    var = np.array(var)
    weights = np.array(weights)
    
    var_2 = var ** 2
    print(var_2)
    weights_2 = weights ** 2
    print(weights_2)

    #Passo 4: Multiplicar a volatilidade ponderada pelo peso de cada ativo
    var_individual = 0
    for i in range(len(assets)):
        var_individual += var_2[i] * weights_2[i]
    
    print(f'Sinaliazar{var_individual}')

    # Adicionar o termo de correlação
    num_assets = len(assets)
    cor = cor_matrix.values
    var_cor = 0
    numero_loops = 0
    for i in range(num_assets):
        for j in range(i + 1, num_assets):
            var_cor += 2 * cor[i, j] * weights[i] * weights[j] * var[i] * var[j]
            numero_loops += 1
    
    print(f'Numero de loops: {numero_loops}')
    print(var_cor)

    # Soma total do risco
    var_total = var_individual + var_cor
    var_total = np.sqrt(var_total)
    print(var_total)

    return var_total

var_tot = calculate_portfolioVaR(df_returns, assets, weights)
```

```python
import numpy as np
from scipy.optimize import minimize

def equations(weights, Mvars, VarPort, Vars, CorrMatrix):
    """
    Retorna um array [eq_1, eq_2, ..., eq_n],
    onde eq_i = w_i - (Mvar_i * VarPort) / sum_j [ w_j * Vars[j] * Vars[i] * CorrMatrix[i, j] ]
    """
    n = len(weights)
    eqs = np.zeros(n)
    for i in range(n):
        denom_i = 0.0
        for j in range(n):
            denom_i += weights[j] * Vars[j] * Vars[i] * CorrMatrix[i, j]

        if abs(denom_i) < 1e-12:
            eqs[i] = 1e6  # penalização grande

        else:
            eqs[i] = weights[i] - (Mvars[i] * VarPort) / denom_i

    return eqs

def objective(weights, Mvars, VarPort, Vars, CorrMatrix):
    """
    Soma dos quadrados das eq_i. (minimizar -> eq_i -> 0)
    """
    eq_vals = equations(weights, Mvars, VarPort, Vars, CorrMatrix)
    return np.sum(eq_vals**2)

def generate_grid_3assets(step=0.1):
    """
    Gera um grid (wA, wB, wC) em passos de 'step' no intervalo [0,1]
    e filtra somente aqueles onde wA + wB + wC <= 1.
    """
    w_range = np.arange(0, 1 + 1e-12, step)
    grid = []
    for wA in w_range:
        for wB in w_range:
            for wC in w_range:
                if wA + wB + wC <= 1.0 + 1e-12:
                    grid.append((wA, wB, wC))
    return grid

if __name__ == "__main__":
    # =======================
    # PARÂMETROS DO PROBLEMA
    # =======================
    n = len(assets)

    # Exemplo de dados
    VarPort = calculate_portfolioVaR(df_returns, assets, weights)
    Mvars   = np.array([0.0035, 0.0004, 0.0002])  # MVAR de cada ativo
    Vars    = np.array([-0.0075, -0.0043, -0.0077])  # Var (ou vol)
    CorrMatrix = np.array([
        [1.0,         0.958242539, -0.270317931],
        [0.958242539, 1.0,         -0.234497113],
        [-0.270317931,-0.234497113, 1.0        ]
    ])

    # =======================
    # GERAÇÃO DA GRADE
    # =======================
    step = 0.1  # Ajuste conforme desejado
    grid = generate_grid_3assets(step=step)

    # =======================
    # AVALIAÇÃO EM TODA A GRADE
    # =======================
    results = []  # para guardar ( (wA, wB, wC), valor_objetivo )

    for (wA, wB, wC) in grid:
        weights = np.array([wA, wB, wC])
        val_obj = objective(weights, Mvars, VarPort, Vars, CorrMatrix)
        results.append((weights, val_obj))

    # Ordenar pelos melhores (menor soma dos quadrados das eqs)
    results.sort(key=lambda x: x[1])

    # =======================
    # IMPRIMIR SÓ OS 3 MELHORES
    # =======================
    
    top_k = 3
    print(f"=== MELHORES {top_k} PONTOS NA GRADE (step={step}) ===")
    for i in range(top_k):
        w_opt, val_obj = results[i]
        sum_w = np.sum(w_opt)
        print(
            f"Rank {i+1:02d}: wA={100*w_opt[0]:.2f}%, "
            f"wB={100*w_opt[1]:.2f}%, "
            f"wC={100*w_opt[2]:.2f}% | "
            f"sum={100*sum_w:.2f}% | "
            f"erro={val_obj:.4e}"
        )
```

```python
# Def Calcular a derivada do portifolio em relação ao Pesoa A

def calculate_weights_constrained(MvarA, MvarB, VarPort, VarA, VarB, Corr, tol=1e-6, max_iter=1000):
    """
    Calcula PesoA e PesoB a partir das equações não-lineares com restrições nos pesos (0.001 <= Peso <= 1).
    
    Parâmetros:
    - MvarA: Margem de VaR do ativo A.
    - MvarB: Margem de VaR do ativo B.
    - VarPort: Variância do portfólio.
    - VarA: VaR do ativo A (pode ser negativo).
    - VarB: VaR do ativo B (pode ser negativo).
    - Corr: Correlação entre os ativos A e B.
    - tol: Tolerância para convergência (padrão: 1e-6).
    - max_iter: Número máximo de iterações (padrão: 1000).
    
    Retorna:
    - PesoA: Peso do ativo A.
    - PesoB: Peso do ativo B.
    """
    # Inicialização dos pesos
    PesoA = 0.5
    PesoB = 0.5

    for iteration in range(max_iter):
        # Calcular novos pesos com base nas equações
        new_PesoA = (MvarA * VarPort) / (PesoA * abs(VarA)**2 + PesoB * abs(VarB) * abs(VarA) * Corr)
        new_PesoB = (MvarB * VarPort) / (PesoB * abs(VarB)**2 + PesoA * abs(VarB) * abs(VarA) * Corr)

        # Garantir que os pesos estão dentro do intervalo [0.001, 1]
        new_PesoA = max(0.001, min(1, new_PesoA))
        new_PesoB = max(0.001, min(1, new_PesoB))

        # Verificar convergência
        if abs(new_PesoA - PesoA) < tol and abs(new_PesoB - PesoB) < tol:
            return new_PesoA, new_PesoB
        # Atualizar os pesos
        PesoA, PesoB = new_PesoA, new_PesoB

    raise ValueError("O método não convergiu dentro do número máximo de iterações.")

# Dados fornecidos
MvarA = 0.00375813
MvarB = 0.001267313
VarPort = 0.005
VarA = -0.0075  # Pode ser negativo
VarB = -0.0043  # Pode ser negativo
Corr = 0.958242539

# Teste da função com os dados fornecidos
try:
    PesoA, PesoB = calculate_weights_constrained(MvarA, MvarB, VarPort, VarA, VarB, Corr)
    PesoA, PesoB
except Exception as e:
    str(e)

print(PesoA, PesoB)
```

```python
import numpy as np
from scipy.optimize import minimize

def equations(weights, Mvars, VarPort, Vars, CorrMatrix):
    """
    Retorna lista eqs = [eq_1, eq_2, ..., eq_n],
    onde eq_i = w_i - (Mvar_i * VarPort)/sum_j(w_j*Vars[j]*Vars[i]*Corr[i,j]).
    """
    n = len(weights)
    eqs = []
    for i in range(n):
        denom_i = 0.0
        for j in range(n):
            denom_i += weights[j] * Vars[j] * Vars[i] * CorrMatrix[i, j]
        # Se o denominador for muito próximo de zero, evite a divisão (pode causar erro numérico)
        if abs(denom_i) < 1e-12:
            eq_i = 1e6  # Penalização gigante
        else:
            eq_i = weights[i] - (Mvars[i] * VarPort) / denom_i
        eqs.append(eq_i)
    return np.array(eqs)

def objective(weights, Mvars, VarPort, Vars, CorrMatrix):
    """
    Função objetivo = soma dos quadrados das eq_i.
    Minimizar essa função faz com que eq_i se aproxime de zero.
    """
    eqs_val = equations(weights, Mvars, VarPort, Vars, CorrMatrix)
    return np.sum(eqs_val**2)

def constraint_sum_weights(weights):
    """
    sum(weights) <= 1  ->  1 - sum(weights) >= 0
    """
    return 1.0 - np.sum(weights)

def positivity_constraint_i(i):
    """
    Peso_i >= 0 -> weights[i] >= 0
    """
    def _con(weights):
        return weights[i]
    return _con

# Exemplo para 3 ativos (A, B, C)
if __name__ == "__main__":

    # Parametros
    VarPort = 0.005
    Mvars = np.array([0.00375813, 0.001267313, 0.0020000])
    Vars  = np.array([-0.0075, -0.0043, -0.0050])
    CorrMatrix = np.array([
        [1.0       , 0.958242539, 0.80],
        [0.958242539, 1.0       , 0.75],
        [0.80      , 0.75      , 1.0 ]
    ])
    
    # Chute inicial
    initial_guess = np.array([0.33, 0.33, 0.34])
    n = len(initial_guess)
    
    # Construir lista de restrições
    # 1) sum(weights) <= 1  => type='ineq'
    # 2) weights[i] >= 0    => type='ineq'
    constraints = [{'type': 'ineq', 'fun': constraint_sum_weights}]
    for i in range(n):
        constraints.append({'type': 'ineq', 'fun': positivity_constraint_i(i)})
    
    # Resolver o problema de minimizacao
    result = minimize(
        fun=objective,
        x0=initial_guess,
        args=(Mvars, VarPort, Vars, CorrMatrix),
        constraints=constraints,
        method='SLSQP',  # 'trust-constr' também é uma opção
        options={'maxiter': 1000, 'disp': True}
    )
    
    # Verificar se convergiu
    if not result.success:
        print("O solver de otimização não convergiu. Mensagem:", result.message)
    
    # Pesos resultado
    w_opt = result.x
    
    # Imprimir resultados
    print("Solução encontrada (minimizando eq_i^2 e impondo sum(w_i) <= 1):")
    for i, w in enumerate(w_opt):
        print(f"Peso {chr(ord('A') + i)}: {w * 100:.4f}%")
    print(f"Soma dos pesos: {np.sum(w_opt)*100:.4f}%")

    # Verificar o valor residual das eq_i
    eq_vals = equations(w_opt, Mvars, VarPort, Vars, CorrMatrix)
    print("eq_i (para cada i):", eq_vals)
    print("Soma dos quadrados das eq_i:", objective(w_opt, Mvars, VarPort, Vars, CorrMatrix))
```
