```python
import os
import pandas as pd

# Caminho da pasta com os arquivos Parquet
#caminho_pasta = r"Z:\Asset Management\Equipe\Emanuel\21.0-Risco\DashRisco\BaseFundos"
caminho_pasta = r"Z:\Asset Management\Equipe\Emanuel\21.0-Risco\DashRisco\BaseFundos"

# Percorre todos os arquivos no diretório
for nome_arquivo in os.listdir(caminho_pasta):
    if nome_arquivo.endswith(".parquet"):
        caminho_parquet = os.path.join(caminho_pasta, nome_arquivo)
        
        # Lê o arquivo parquet
        df = pd.read_parquet(caminho_parquet)
        
            # Define o nome do arquivo de saída .csv
            nome_csv = nome_arquivo.replace(".parquet", ".csv")
            caminho_csv = os.path.join(caminho_pasta, nome_csv)
            
            # Salva como CSV
            df.to_csv(caminho_csv, index=False)
            print(f"Convertido: {nome_arquivo} -> {nome_csv}")
```
